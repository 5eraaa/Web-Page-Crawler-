```python
import string
```

```python
def load_stop_words(file_path):
    try:
        with open(file_path, 'r') as file:
            words_string = file.read()
            words_list = words_string.split(',')
            stop_words = {word.strip() for word in words_list}
        return stop_words
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return set()
```

```python
def remove_stop_words_and_punctuation(text, stop_words):
    text = text.lower()
    text = text.translate(str.maketrans('', '',string.punctuation))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return filtered_words
```

```python
#Bubble sort
def bubble_sort(l):
    bubble=False
    for i in range(len(l)-1,0,-1):
        if bubble :
            return l
        bubble=True
        for j in range(i):
            if l[j][1]<l[j+1][1]:
                l[j],l[j+1]=l[j+1],l[j]
                bubble=False
```

```python
#insertion sort

def insertion_sort(l):
    for i in range (1,len(l)):
        pos=i
        value=l[i]
        while pos>0 and l[pos-1][1]<value[1]:
            l[pos]=l[pos-1]
            pos-=1

        l[pos]=value


    return l
```

```python
#Quick sort
def partition(alist, first, last):
    pivotvalue = alist[first]
    leftmark = first + 1
    rightmark = last
    done = False

    while not done:
        while leftmark <= rightmark and alist[leftmark][1] >= pivotvalue[1]:
            leftmark += 1

        while rightmark >= leftmark and alist[rightmark][1] <= pivotvalue[1]:
            rightmark -= 1

        if rightmark < leftmark:
            done = True
        else:
            alist[leftmark], alist[rightmark] = alist[rightmark], alist[leftmark]

    alist[first], alist[rightmark] = alist[rightmark], alist[first]

    return rightmark

def quickSortHelper(alist, first, last):
    if first < last:
        splitpoint = partition(alist, first, last)
        quickSortHelper(alist, first, splitpoint - 1)
        quickSortHelper(alist, splitpoint + 1, last)

def quick_sort(alist):
    quickSortHelper(alist, 0, len(alist) - 1)
    return alist
```

```python
#Merge sort
def merge(items, left, mid, right):
    n1 = mid - left + 1
    n2 = right - mid

    leftitem= items[left:left + n1]
    rightitem = items[mid + 1:mid + 1 + n2]

    i = j = 0
    k = left

    while i < n1 and j < n2:
        if leftitem[i][1] >= rightitem[j][1]:
            items[k] = leftitem[i]
            i += 1
        else:
            items[k] = rightitem[j]
            j += 1
        k += 1

    while i < n1:
        items[k]= leftitem[i]
        i += 1
        k += 1

    while j < n2:
        items[k] = rightitem[j]
        j += 1
        k += 1


def merge_sort(items):
    merge_sort_helper(items, 0, len(items)-1)
    return items


def merge_sort_helper(items, left, right):
    if left < right:
        mid = (left + right) // 2

        merge_sort_helper(items, left, mid)
        merge_sort_helper(items, mid + 1, right)
        merge(items, left, mid, right)
```

```python
#Selection sort
def selection_sort(items):

    n = len(items)


    for i in range(n-1):
        max=items[i][1]
        max_idx = i
        for j in range(i+1, n):
            if items[j][1] > max:
                max=items[j][1]
                max_idx = j
        items[i], items[max_idx] = items[max_idx],items[i]
    return items
```

```python
def classify_text(text,k):
    stoping_words_file_path='/content/common-english-words.txt'
    stop_words = load_stop_words(stoping_words_file_path)

    if not stop_words:
        return "Error: Stop words file not loaded properly."

    cleaned_text = remove_stop_words_and_punctuation(text, stop_words)

    unique_elements = set(cleaned_text)
    redundances = [(item, cleaned_text.count(item)) for item in unique_elements]

    print("Choose the sorting algorithm:\n1 - Bubble sort\n2 - Insertion sort\n3 - Quick sort\n4 - Merge sort\n5 - Selection sort")

    try:
        choice = int(input("Enter your choice (1-5): "))
    except ValueError:
        return "Invalid input. Please enter a number between 1 and 5."

    if choice == 1:
        sorted_items = bubble_sort(redundances)
    elif choice == 2:
        sorted_items = insertion_sort(redundances)
    elif choice == 3:
        sorted_items = quick_sort(redundances)
    elif choice == 4:
        sorted_items = merge_sort(redundances)
    elif choice == 5:
        sorted_items = selection_sort(redundances)
    else:
        return "Invalid choice. Please enter a number between 1 and 5."

    top_words = [item[0] for item in sorted_items[:k+1]]

    if any(word in top_words for word in ["football", "ball", "match"]):
        category = "Football"
    elif any(word in top_words for word in ["economy", "budget", "market"]):
        category = "Economy"
    elif any(word in top_words for word in ["fashion", "latest", "outfit"]):
        category = "Fashion"
    elif any(word in top_words for word in ["poetry", "art", "music"]):
        category = "Art"
    else:
        category = "Unknown"

    return f"Category: {category}"
```

```python
user_text = input("Please enter the text to classify: ")
k=input("Enter number of words to classify based on it : ");
result = classify_text(user_text,k)
print(result)
```
